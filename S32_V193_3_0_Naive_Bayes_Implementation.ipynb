## Naive Baye's Machine Learning Algorithms Implementation

```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
X,y=load_iris(return_X_y=True)
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
y_pred=gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
df['Sex_encoded']=lb.fit_transform(df['sex'])
df['smoker_encoded']=lb.fit_transform(df['smoker'])
df['day_encoded']=lb.fit_transform(df['day'])
df['time_encoded']=lb.fit_transform(df['time'])
```

```python
df.columns
```

```python
X=df[['total_bill', 'tip', 'Sex_encoded', 'smoker_encoded', 'day_encoded']]
y=df['time_encoded']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_test
```

```python
sc=StandardScaler()
X_train[['total_bill', 'tip']]=sc.fit_transform(X_train[['total_bill', 'tip']])
X_test[['total_bill', 'tip']]=sc.transform(X_test[['total_bill', 'tip']])
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
y_pred=gnb.fit(X_train,y_train).predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```
